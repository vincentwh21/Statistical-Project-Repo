```python
import pandas as pd
import numpy as np
```

```python
order_items = pd.read_csv('dataset/olist_order_items_dataset.csv')
orders = pd.read_csv('dataset/olist_orders_dataset.csv')

order_items = order_items.drop_duplicates(subset='order_id')
orders = orders.drop_duplicates(subset='order_id')

# 通过订单号合并买家和卖家
seller_customer = pd.merge(order_items, orders, on='order_id', how='inner')
# seller_customer = seller_customer[['order_id', 'customer_id', 'seller_id']]
# 筛选出时间shipping_limit_date为2018年6月的数据
seller_customer = seller_customer[seller_customer['shipping_limit_date'].str.contains('2018-06')]
seller_customer
```

```python
customers = pd.read_csv('dataset/olist_customers_dataset.csv')
sellers = pd.read_csv('dataset/olist_sellers_dataset.csv')
geolocation = pd.read_csv('dataset/olist_geolocation_dataset.csv')

seller_customer = pd.merge(seller_customer, customers[['customer_id', 'customer_zip_code_prefix']], on='customer_id', how='left')
seller_customer = pd.merge(seller_customer, sellers[['seller_id', 'seller_zip_code_prefix']], on='seller_id', how='left')

seller_customer
```

```python
geolocation.drop_duplicates(subset='geolocation_zip_code_prefix', inplace=True)
geolocation
```

```python
# 合并卖家数据与地理位置信息
merged_data_seller = pd.merge(
    seller_customer, 
    geolocation[['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']], 
    left_on='seller_zip_code_prefix', 
    right_on='geolocation_zip_code_prefix', 
    how='left'
)

# 重命名合并后的列，以区分这些列属于卖家
merged_data_seller.rename(columns={
    'geolocation_zip_code_prefix': 'seller_geo_zip_prefix',
    'geolocation_lat': 'seller_geo_lat',
    'geolocation_lng': 'seller_geo_lng'
}, inplace=True)
```

```python
# 将买家的经纬度信息连接到买家数据
merged_data = pd.merge(
    merged_data_seller, 
    geolocation[['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']], 
    left_on='customer_zip_code_prefix', 
    right_on='geolocation_zip_code_prefix', 
    how='left'
)

# 重命名合并后的列，以区分这些列属于买家
merged_data.rename(columns={
    'geolocation_zip_code_prefix': 'buyer_geo_zip_prefix',
    'geolocation_lat': 'buyer_geo_lat',
    'geolocation_lng': 'buyer_geo_lng'
}, inplace=True)
```

```python
merged_data.dropna(inplace=True)
merged_data
```

```python
import folium

# 创建一个世界地图
m = folium.Map(location=[20, 0], zoom_start=2)

# 遍历数据框中的每一行
for idx, row in merged_data.iterrows():
    # 定义买家和卖家的位置
    seller_location = [row['seller_geo_lat'], row['seller_geo_lng']]
    buyer_location = [row['buyer_geo_lat'], row['buyer_geo_lng']]
    
    # # 添加卖家位置标记
    # folium.Marker(
    #     seller_location,
    #     popup='Seller',
    #     icon=folium.Icon(color='red', icon='info-sign')
    # ).add_to(m)

    # # 添加买家位置标记
    # folium.Marker(
    #     buyer_location,
    #     popup='Buyer',
    #     icon=folium.Icon(color='blue', icon='info-sign')
    # ).add_to(m)

    # 在买家和卖家之间画线
    folium.PolyLine(
        locations=[buyer_location, seller_location],
        color='green',
        weight=0.1,
        opacity=0.8
    ).add_to(m)

# 显示地图
m.save('seller_buyer_map.html')
```

```python
import folium

# 创建一个世界地图
m = folium.Map(location=[20, 0], zoom_start=2, tiles='Stamen Terrain')

# 遍历数据框中的每一行
for idx, row in merged_data.iterrows():
    seller_location = [row['seller_geo_lat'], row['seller_geo_lng']]
    buyer_location = [row['buyer_geo_lat'], row['buyer_geo_lng']]
    
    # 在地图上添加卖家位置标记
    folium.CircleMarker(
        seller_location,
        radius=0.01,
        color='red',
        fill=True,
        fill_color='red',
        popup='Seller'
    ).add_to(m)

    # 在地图上添加买家位置标记
    folium.CircleMarker(
        buyer_location,
        radius=0.01,
        color='blue',
        fill=True,
        fill_color='blue',
        popup='Buyer'
    ).add_to(m)

    # 在买家和卖家之间画线
    folium.PolyLine(
        locations=[buyer_location, seller_location],
        color='green',
        weight=0.2,
        opacity=0.8
    ).add_to(m)

# 显示地图
m.save('map.html')
```
