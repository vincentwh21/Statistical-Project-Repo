```python
wt = 1 / df.review_score.value_counts(normalize= True)
df['weight'] = df.review_score.apply(lambda x: wt[x])

avg_reviews = np.mean(df[['product_id','review_score']].groupby('product_id').count().review_score)

print(wt)
print(avg_reviews)
```

```python
def scoring(group):
    n_reviews = group.count().review_score
    score = np.average(group.review_score, weights= group.weight**2) * (n_reviews/avg_reviews)
    return score
    
scores_overall = df.groupby('product_id').apply(scoring)
scores_overall.head(5)
```

```python
cols= [
    'product_id',
    'order_purchase_timestamp',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'order_delivered_carrier_date',
    'shipping_limit_date'
]

prod_delivery = pd.merge(
    left= orders,
    right= ord_items,
    how= 'inner',
    on= 'order_id'
)[cols]

date_cols = [
    'order_purchase_timestamp',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'order_delivered_carrier_date',
    'shipping_limit_date'
]
for col in date_cols:
    prod_delivery[col]= pd.to_datetime(prod_delivery[col])

prod_delivery.head(5)
```

```python
def delivery_performance(est_date, del_date):
    if est_date > del_date:
        return "Early"
    elif est_date < del_date:
        return "Late"
    else:
        return "On time"
    

def delivery_days(ord_date, del_date):
    return (del_date - ord_date).days


def ship_limit(car_date, ship_limit):
    if car_date > ship_limit:
        return "Limit crossed"
    else:
        return "Within limits"

    
prod_delivery['delivery_perf'] = prod_delivery.apply(
    lambda x: delivery_performance(x.order_estimated_delivery_date, x.order_delivered_customer_date), 
    axis=1
)


prod_delivery['delivery_days'] = prod_delivery.apply(
    lambda x: delivery_days(x.order_purchase_timestamp, x.order_delivered_customer_date),
    axis= 1
)

    
prod_delivery['shipping_limit'] = prod_delivery.apply(
    lambda x: ship_limit(x.order_delivered_carrier_date, x.shipping_limit_date),
    axis= 1
)
```

```python
top_50 = scores_overall.sort_values(ascending= False)[:50]
top_50_data = prod_delivery.set_index('product_id').loc[top_50.index]
top_50_data.sample(10)
```

```python
prod_df = pd.merge(
    left= df,
    right= products,
    on= 'product_id',
    how= 'left'
)
prod_df
```

```python
cols= [
    'product_id',
    'product_category_name_english',
    'review_score'
]

prod_catg_scores = prod_df[cols]
prod_catg_scores
```

```python
prod_catg_scores['weight'] = 0

tmp = prod_catg_scores.groupby(
    'product_category_name_english'
).apply(
    lambda x: 1 / x.review_score.value_counts(normalize=True)
).reset_index()

for ind,row in tmp.iterrows():
    prod_catg_scores.loc[
        (
            prod_catg_scores.review_score == row.level_1
        ) & (
            prod_catg_scores.product_category_name_english == row.product_category_name_english
        ),
        'weight'
    ] = row.review_score

prod_catg_scores
```

```python
prod_catg_scores.dropna(subset= ['product_category_name_english'], inplace= True)
categories = prod_catg_scores.product_category_name_english.unique()

for catg in categories:
    data = prod_catg_scores.loc[prod_catg_scores.product_category_name_english == catg,['product_id','review_score','weight']].groupby('product_id')
    avg_reviews = np.mean(data.count().review_score)
    
    scores_catg = data.apply(scoring).sort_values(ascending= False)
    print(catg)
    print(scores_catg[:min(5,len(scores_catg))])
    print("\n")
```
